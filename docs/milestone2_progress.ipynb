**Progress Report for Team 11 - 4th Prime**

Currently, we have our progress broken up into two main areas of focus. We have one half of the team working on implementing and validating our dual numbers class, and the other half is working on implementing and validating our derivative definitions. We deliberately decided to have the person making the test cases separate from the person implementing the functions, so that our tests accurately reflect our desired functionality without any biases from the specifications of the implementation. In addition, all group members are working on building up the documentation since milestone 1.
